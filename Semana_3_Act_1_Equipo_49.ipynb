**Parte 1: Fundamentos de bases de datos y almacenes de datos (Data Warehouse) para ciencia de datos.**


Sabemos que la información es un activo importante en las empresas y por ello se procura que esta se encuentre organizada y concentrada en una base de datos.

Una **base de datos** es una colección organizada de datos lógicamente relacionados. Funciona como un modelo particular del mundo real en forma de datos. Supongamos una base de datos en una escuela; tendrá información de los profesores, de los alumnos, de las materias, las calificaciones, los horarios con lo que se puede mostrar la estructura total.

Las bases de datos pueden tener una o más tablas.
Una **tabla** es un concentrado de datos organizados en filas y columnas
Las **filas** también son llamadas registros y las columnas campos.
Una **celda** es la intersección entre filas y columnas.
Los **datos** se refieren al conjunto de información de todos los registros más los campos.
Un **objeto** es una entidad que en la que varían las formas de imprimir o mostrar la información.

Un **Sistema de administración de base de datos** es un software que organiza la información, proporciona accesos y existen lenguajes para su manipulación como SQL y DML.

Este enfoque proporciona ventajas como reducir la redundancia de datos, mejora la consistencia, mejora el intercambio de datos, incrementa la calidad de datos, facilita el acceso y descarga de información.

**Data Warehouse** es un almacén electrónico de gran cantidad de información de distintas fuentes.

```python
# Eliminar las filas donde todos los elementos son nulos
df =  pd.read_csv(c)
df.index.name = None
print("Antes " + str(len(df)))
df.dropna(how='all', inplace = True)
print("Después " + str(len(df)))

# Se observa que ninguna fila tiene todos los elemtos nulos
```

```python
# Se puede usar Threshold para mantener las filas con 
# Al menos 5 valores nulos

n = 5
df =  pd.read_csv(c)
df.index.name = None
print("Antes " + str(len(df)))
df.dropna(thresh= n  , inplace = True)
print("Después " + str(len(df)))

```

**Solución 2 :**

```python
# Remplazar el campo nulo con el promedio

df =  pd.read_csv(c)
df.index.name = None

df_ant=df.copy()

wm = df_ant.X5.mean()
wm

df_ant.isnull().values.any()
```

```python
# Traes asignar un valor ya no tenemos valores nulos
# En este caso fue la media

df_ant.fillna(value = wm,
                 inplace = True)

df_ant.isnull().values.any()
```

```python
# Pero la media puede estar muy sesgada y por ello, 
# vale la pena considerar  la moda o la mediana :

#  moda = df_ant.X5.mode()
#  mediana = df_ant.X5.median()

```

```python
# Es posible eliminar las filas donde las columnas
# señaladas tengan un dato nulo


print(df.isnull().values.any())
df.dropna(subset=['X3', 'X4'], inplace = True)

# Aquí X3 ya no tiene valores nulos
print(df['X3'].isnull().values.any())
```

```python
# Es posible asignar a cada columna una estrategia distinta
df.X7.fillna(df.X7.mode()[0], inplace=True)
df.X8.fillna(df.X8.mean(), inplace=True)
df[['X7','X8']].tail()
```

```python
#Para conocer los estadísticos de la base
df.describe()
# Esto nos da visión para elegir entre media y mediana
```
